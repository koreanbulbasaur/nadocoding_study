# 16. 이미지 변환 (열림 & 닫힘)
## 열림 (Opening) : 침식 후 팽창, 깎아서 노이즈 제거 후 살 찌움
> dilate(erode(image))

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = np.ones((3, 3), np.uint8)
n = 4

img = cv2.imread('./img/paint2.png', 0)

erode = cv2.erode(img, kernel, iterations=n)
dilate = cv2.dilate(erode, kernel, iterations=n)

cv2.imshow('img', img)
cv2.imshow('erode', erode)
cv2.imshow('dilate', dilate)

cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

lst = [img, erode, dilate]
lst_t = ['img', 'erode', 'dilate']
idx = 0

for x in range(2):
    for y in range(2):
        if x == 1 and y == 1:
            ax[x, y].axis('off')
        else:
            ax[x, y].imshow(lst[idx], cmap='gray')
            ax[x, y].axis('off')
            ax[x, y].set_title(lst_t[idx])
            idx += 1
```

## 닫힘 (Closing) : 팽창 후 침식. 구멍을 메운 후 다시 깎음
> erode(dilate(image))

```python
kernel = np.ones((3, 3), np.uint8)
n = 4

img = cv2.imread('./img/paint1.png', 0)

dilate = cv2.dilate(img, kernel, iterations=n)
erode = cv2.erode(dilate, kernel, iterations=n)

cv2.imshow('img', img)
cv2.imshow('dilate_1', dilate)
cv2.imshow('erode_2', erode)

cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

lst = [img, dilate, erode]
lst_t = ['img', 'dilate', 'erode']
idx = 0

for x in range(2):
    for y in range(2):
        if x == 1 and y == 1:
            ax[x, y].axis('off')
        else:
            ax[x, y].imshow(lst[idx], cmap='gray')
            ax[x, y].axis('off')
            ax[x, y].set_title(lst_t[idx])
            idx += 1
```
