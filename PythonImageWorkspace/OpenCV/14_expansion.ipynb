# 14. 이미지 변환 (팽창)

## 이미지를 확장하여 작은 구멍을 채움
흰색 영역의 외곽 픽셀 주변에 흰색을 추가

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

kernel = np.ones((3, 3), np.uint8)
kernel
```

```python
img = cv2.imread('./img/paint1.png', 0)
dilate1 = cv2.dilate(img, kernel, iterations=1, )
dilate2 = cv2.dilate(img, kernel, iterations=2, )
dilate3 = cv2.dilate(img, kernel, iterations=3, )

cv2.imshow('gray', img)
cv2.imshow('dilate1', dilate1)
cv2.imshow('dilate2', dilate2)
cv2.imshow('dilate3', dilate3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

lst = [img, dilate1, dilate2, dilate3]
lst_t = ['img', 'dilate1', 'dilate2', 'dilate3']
idx = 0

for x in range(2):
    for y in range(2):
        ax[x, y].imshow(lst[idx], cmap='gray')
        ax[x, y].axis('off')
        ax[x, y].set_title(lst_t[idx])
        idx += 1
```
