# 13. 이미지 변형 (이진화)

```python
import cv2
img = cv2.imread('./img/book.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (640, 960))

ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.namedWindow('img', cv2.WINDOW_NORMAL)
cv2.imshow('img', img)
cv2.imshow('binary', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Trackbar (값 변화에 따른 변형 확인)

```python
import cv2

def empty(pos):
    print(pos)
    pass

img = cv2.imread('./img/book.jpg', cv2.IMREAD_GRAYSCALE)
# img = cv2.resize(img, (960, 640))

name = 'Trackbar'
cv2.namedWindow(name, cv2.WINDOW_NORMAL)

cv2.createTrackbar('threshold', name, 127, 255, empty) # bar 이름, 창의 이름, 초기값, 최대값, 이벤트 처리

while True:
    thresh = cv2.getTrackbarPos('threshold', name)

    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)

    if not ret:
        break

    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```
