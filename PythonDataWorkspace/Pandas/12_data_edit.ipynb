# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['club'].replace({'council':'EN', 'GM': 'JP_GM'})
```

```python
df
```

```python
df['club'].replace({'council':'EN'}, inplace=True)
df
```

```python
df['SW'].str.lower()
```

```python
df['SW'] = df['SW'].str.lower()
df
```

```python
df['SW'] = df['SW'].str.upper()
df
```

```python
df['club'] = 'Hololive_'+ df['club']
df
```

## Column 추가

```python
df['total'] = df['Language'] + df['English'] + df['Math'] + df['science'] + df['Society']
df
```

```python
df['total'] = 'Fail' # 결과 Column 을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
df.loc[df['total'] >300, 'result'] = 'Pass' # 총합이 300 보다 큰 데이터에 대해서 결과를 Pass로 업데이트 
df
```

## Column 삭제

```python
df.drop(columns=['total']) # 총합 column 삭제
```

```python
df.drop(columns=['Language', 'English', 'Math'], inplace=True)
```

```python
df
```

## Row 삭제

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df.drop(index='4번') # 4번 학생 데이터 row 를 삭제
```

## Cell 수정

```python
df.loc['4번', 'SW'] = 'Python' # 4번 학생의 SW특기 데이터를 Python 으로 변경
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
# 맨 뒤에 있는 결과 Column 을 앞으로 가져오고, 나머지 Column 들과 합쳐서 순서 변경
df = df[[cols[-1]] + cols[0:-1]]
df
```
