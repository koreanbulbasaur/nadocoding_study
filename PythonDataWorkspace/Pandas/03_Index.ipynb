```python
import pandas as pd
import numpy as np
from random import *
```

```python
degree = [i for i in range(10, 101, 5)]
data = {
    'name' : ['kronii', 'botan', 'anya', 'mio', 'roboco', 'Marine', 'Baelz', 'okayu'],
    'club' : ['council', 'JP5', 'ID2', 'GM', 'JP0', 'JP3', 'council', 'GM'],
    'height' : [168, 166, 147, 160, 154, 150, 149, 152],
    'Language' : sample(degree, 8),
    'English' : sample(degree, 8),
    'Math' : sample(degree, 8),
    'science' : sample(degree, 8),
    'Society' : sample(degree, 8),
    'SW' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번',
                               '7번', '8번'])
df
```

```python
df.index
```

### Index 이름 설정

```python
df.index.name = '지원번호'
df
```

## Index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop= True) # 원래 쓰던 '지원번호' : 인덱스 삭제
```

```python
df.reset_index(drop=True, inplace= True) # 실제 데이터에 바로 반영
df
```

## Index 설정
지정한 column 으로 Index 를 설정

```python
df.set_index('name')
```

## Index 정렬
Index 를 기준으로 오름차순, 내림차순 정렬

```python
df.sort_index() # 인덱스로 오름차순 정렬
```

```python
df.sort_index(ascending=False) # 내림차순으로 정렬
```
