## Data 준비
사전 (dict) 자료 구조를 통해 생성
예) 슬램덩크 주요 인물 8명에 대한 데이터

```python
from random import *
degree = [i for i in range(10, 101, 5)]
```

```python
data = {
    'name' : ['kronii', 'botan', 'anya', 'mio', 'roboco', 'Marine', 'Baelz', 'okayu'],
    'club' : ['council', 'JP5', 'ID2', 'GM', 'JP0', 'JP3', 'council', 'GM'],
    'height' : [168, 166, 147, 160, 154, 150, 149, 152],
    'Language' : sample(degree, 8),
    'English' : sample(degree, 8),
    'Math' : sample(degree, 8),
    'science' : sample(degree, 8),
    'Society' : sample(degree, 8),
    'SW' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
data
```

```python
data['name']
```

```python
data['height']
```

## DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

### 데이터 접근

```python
df['name']
```

```python
df['height']
```

```python
df[['name', 'height']]
```

## DataFrame 객체 생성(Index 지정)

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번',
                               '7번', '8번'])
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column 만 선택하거나, 순서 변경 가능

```python
df = pd.DataFrame(data, columns=['name', 'club', 'height'])
df
```

```python
df = pd.DataFrame(data, columns=['name', 'height', 'club'])
df
```
