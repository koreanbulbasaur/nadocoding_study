# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd
from random import *
degree = [i for i in range(10, 101, 5)]
data = {
    'name' : ['kronii', 'botan', 'anya', 'mio', 'roboco', 'Marine', 'Baelz', 'okayu'],
    'club' : ['council', 'JP5', 'ID2', 'GM', 'JP0', 'JP3', 'council', 'GM'],
    'height' : [168, 166, 147, 160, 154, 150, 149, 152],
    'Language' : sample(degree, 8),
    'English' : sample(degree, 8),
    'Math' : sample(degree, 8),
    'science' : sample(degree, 8),
    'Society' : sample(degree, 8),
    'SW' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번',
                               '7번', '8번'])
df.index.name = '지원번호'
df
```

## 저장하기
### csv 파일로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig')
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False)
```

## 텍스트(.txt) 파일로 저장

```python
df.to_csv('score.txt', sep='\t') # tab 으로 구분된 텍스트 파일
```

## 엑셀 파일로 저장

## 열기
### csv 파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 갯수 만큼의 row 를 건너뜀
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1, 3, 5]) # 1, 3, 5 row 는 제외 (row 0부터 시작)
df
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 갯수 만큼의 row 만 가져옴
df
```

```python
# 처음 2 row 무시, 이후에 4 row를 가져옴
df = pd.read_csv('score.csv', skiprows=2, nrows=4) 
df
```

## 텍스트(.txt) 파일 열기

```python
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

```python
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace= True)
df
```
