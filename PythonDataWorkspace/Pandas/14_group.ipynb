# 14. 그룹화

동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용


```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('club')
```

```python
df.groupby('club').get_group('JP')
```

```python
df.info()
```

```python
df.groupby(['club'],as_index=False).sum()
```

```python
df = df.drop('name',axis=1)
df
```

```python
df = df.drop('SW', axis=1)
df
```

```python
df.groupby('club').mean()
```

```python
df.groupby('club').size() # 각 그룹의 크기
```

```python
df.groupby('club').size()['EN']
```

```python
df.groupby('club')['height'].mean()
```

```python
df.groupby('club')[['Language', 'English', 'Math']].mean()
```

```python
df['grade'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
df.groupby(['club','grade']).mean()
```

```python
df.groupby('club').mean()
```

```python
df.groupby('club').mean().sort_values('height')
```

```python
df.groupby('club').mean().sort_values('height', ascending=False)
```

```python
df.groupby(['club', 'grade']).sum()
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df.groupby('club')['SW'].count()
```

```python
df.groupby('club')[['SW', 'name']].count()
```

```python
df['grade'] = [3, 3, 2, 1, 1, 3, 2, 2]
club = df.groupby('club')
club['grade'].value_counts()
```

```python
club['grade'].value_counts().loc['JP']
```

```python
club['grade'].value_counts(normalize=True).loc['JP']
```
