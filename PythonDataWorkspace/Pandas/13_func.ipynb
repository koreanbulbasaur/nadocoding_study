## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm 을 붙이는 역할
def add_cm(height):
    return str(height) + ' cm'

# 키 데이터에 대해서 add_cm 함수로 호출한 결과 데이터를 반영
df['height'] = df['height'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN이 아닌지
        return lang.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW'] = df['SW'].apply(capitalize)
df
```
