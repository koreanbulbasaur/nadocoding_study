# 데이터 정렬

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('height') # 키 기준으로 오름차순 정렬
```

```python
df.sort_values('height', ascending=False) # 키 기준으로 내림차순 정렬
```

```python
df.sort_values(['Math', 'English'], ascending=False)
```

```python
df.sort_values(['Math', 'English']) # 수학, 영어 점수 기준으로 오름차순
```

```python
df.sort_values(['Math', 'English'], ascending=False) # 수학, 영어 점수 기준으로 내름차순
```

```python
df.sort_values(['Math', 'English'],ascending=[True, False])
# 수학 점수는 오름차순으로, 영어 점수는 내림차순으로 정렬
```

```python
df['height'].sort_values()
```

```python
df['height'].sort_values(ascending=False)
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```
