# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['height'] >= 160 # 학생들의 키가 185 이상인지 여부를 True / False
```

```python
filt = (df['height'] >= 160)
df[filt]
```

```python
df[~filt] # filter를 역으로 적용
```

```python
df[df['height']>= 160] 
```

```python
df.loc[df['height'] >= 160,'Math'] # 키가 160 이상인 멤버들의 수학 데이터
```

```python
df.loc[df['height']>=160,['name', 'Math','science']]
```

##  다양한 조건

### & 그리고

```python
# 키가 185 이상인 북산고 학생 데이터
df.loc[(df['height'] >= 160) & (df['club'] == 'council')]
```

| 또는

```python
df.loc[(df['height'] > 160) | (df['height']< 150)]
```

### str 함수

```python
filt = df['name'].str.startswith('k') # 이름이 k 로 시작하는 멤버
df[filt]
```

```python
filt = df['name'].str.contains('o') # 이름에 o 가 들어가는 멤버
df[filt]
```

```python
df[-filt] # 이름에 o 가 안 들어간 멤버
```

```python
langs = ['Python', 'Java']
filt = df['SW'].isin(langs) # sw가 python 이거나 java 인 사람
df[filt] 
```

```python
langs = ['python', 'java']
filt = df['SW'].str.lower().isin(langs)
df[filt]
```

```python
df['SW'].str.contains('Java', na=True) # NaN 데이터에 대해서 True 로 설정
```

```python
df['SW'].str.contains('Java', na=False) # NaN 데이터에 대해서 False 로 설정
```

```python
filt = df['SW'].str.contains('Java', na=False)
df[filt]
```
