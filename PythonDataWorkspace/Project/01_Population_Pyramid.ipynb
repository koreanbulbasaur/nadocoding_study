# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
filename = '201301_201304_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(filename, skiprows=3, 
                     index_col='행정기관', usecols='B,E:Y')
df_m.head()
```

```python
df_m.iloc[0]
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(filename, skiprows=3, 
                     index_col='행정기관', usecols='B,AB:AV')
df_w.head()
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

## 데이터 시각화

```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
plt.barh(df_m.columns, df_m.iloc[0])
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2013 대한민국 인구 피라미드')
plt.savefig('2013_인구피라미드.png',dpi=100)
plt.show()
```
