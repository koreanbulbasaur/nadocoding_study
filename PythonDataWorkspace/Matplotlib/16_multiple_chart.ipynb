# 16. 여러 그래프

```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False
import numpy as np
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

axs[0, 0].bar(df['name'], df['Language'], label='Language_score')
axs[0, 0].set_title('First_Chart')
axs[0, 0].legend()
axs[0, 0].set(xlabel='name', ylabel='score')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['name'], df['Language'], label='Language_score')
axs[0, 0].set_title('First_Chart')
axs[0, 0].legend()
axs[0, 0].set(xlabel='name', ylabel='score')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['name'], df['Math'], label='Math')
axs[0, 1].plot(df['name'], df['English'], label='English')
axs[0, 1].legend()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['name'], df['Language'], label='Language_score')
axs[0, 0].set_title('First_Chart')
axs[0, 0].legend()
axs[0, 0].set(xlabel='name', ylabel='score')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['name'], df['Math'], label='Math')
axs[0, 1].plot(df['name'], df['English'], label='English')
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['name'],df['height'])

# 네 번째 그래프
axs[1, 1].plot(df['name'], df['Society'], color='g', alpha=0.5)
```
