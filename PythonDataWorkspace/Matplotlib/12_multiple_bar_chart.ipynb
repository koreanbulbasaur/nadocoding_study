```python
N = df.shape[0]
N
```

```python
index = np.arange(N)
index
```

```python
w = 0.25
plt.bar(index - w, df['Language'])
```

```python
w = 0.25
plt.bar(index - w, df['Language'])
plt.bar(index, df['English'])
plt.bar(index + w, df['Math'])
```

```python
w = 0.25
plt.bar(index - w, df['Language'], width=w)
plt.bar(index, df['English'], width=w)
plt.bar(index + w, df['Math'], width=w)
```

```python
w = 0.25
plt.bar(index - w, df['Language'], width=w, label='Language')
plt.bar(index, df['English'], width=w, label='English')
plt.bar(index + w, df['Math'], width=w, label='Math')
plt.legend()
```

```python
w = 0.25
plt.bar(index - w, df['Language'], width=w, label='Language')
plt.bar(index, df['English'], width=w, label='English')
plt.bar(index + w, df['Math'], width=w, label='Math')
plt.legend(ncol=3)
```

```python
plt.figure(figsize=(10, 5))
plt.title('학생별 성적')

w = 0.25
plt.bar(index - w, df['Language'], width=w, label='Language')
plt.bar(index, df['English'], width=w, label='English')
plt.bar(index + w, df['Math'], width=w, label='Math')
plt.legend(ncol=3)
plt.xticks(index, df['name'], rotation=60)
plt.show()
```
