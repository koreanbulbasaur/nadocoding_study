# 15. 산점도 그래프

```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus']=False
import numpy as np
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['grade'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['English'], df['Math'])
plt.xlabel('English score')
plt.ylabel('Math score')
```

```python
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['English'], df['Math'], s=sizes)
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
sizes = df['grade'] * 500 # 1 학년 = 500, 2학년 = 1000, 3학년 = 1500
plt.scatter(df['English'], df['Math'], s=sizes)
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis')
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis',
            alpha=0.3)
plt.xlabel('English Score')
plt.ylabel('Math Score')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis',
            alpha=0.3)
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar()
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis',
            alpha=0.3)
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar(ticks=[1, 2, 3], label='grade', shrink=0.3)
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['English'], df['Math'], s=sizes, c=df['grade'], cmap='viridis',
            alpha=0.3)
plt.xlabel('English Score')
plt.ylabel('Math Score')
plt.colorbar(ticks=[1, 2, 3], label='grade', shrink=0.5, 
             orientation = 'horizontal')
```
